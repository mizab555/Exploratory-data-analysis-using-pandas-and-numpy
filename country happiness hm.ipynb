```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
```

```python
aaa=pd.read_csv('World_Happiness_2015.csv')
aaa
```

```python
grp=aaa.groupby('Region')
grp.get_group('Sub-Saharan Africa')
```

```python
grp.mean()
```

```python
grp.max()
```

```python
grp.max()-grp.mean()
```

```python
sampl_mean=grp['Happiness Score'].mean()
sampl_mean
```

```python
ppl_mean=aaa['Happiness Score'].mean()
```

```python
sampling_error=sampl_mean-ppl_mean
sampling_error
```

```python
barh=aaa['Happiness Score']
barh
```

```python
barp=aaa['Region'].value_counts()
barp=aaa['Region'].sort_index()
barp.shape
```

```python
barp
```

```python
figure,ax=plt.subplots()
ax.bar(barp,barh)
plt.xticks(rotation=45)
plt.show() 
```
